# Linear search

```python
# generate random numbers
import numpy as np
n=20000000
X=np.random.randint(0,10,(n,))
print(X)
```

```python
%%time

def linear_search(X, target_value): # O(n)
  found=False
  for i in range(len(X)):
    if X[i]==target_value:
      found=True
  return found


res=linear_search(X, 1)
print(res)
```

```python
%%time

def linear_search(X, target_value): # O(n)
  for i in range(len(X)):
    if X[i]==target_value:
      return True
  return False


res=linear_search(X, 1)

print(res)
```

# Scaling numbers

```python
%%time

Xf=np.asarray(X,np.float32) # O(n) converting from int32 -> float32
minimum=min(Xf) # O(n)
maximum=max(Xf) # O(n)
for i in range(len(X)): # O(n)
    Xf[i]=float(Xf[i]-minimum)/float(maximum-minimum)
    # Xf is [0,1]
print(Xf)


# O(n)
```
