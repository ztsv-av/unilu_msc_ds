# Sliding window

O(n*m) with `n` data samples, and `m` the window size



```python
import numpy as np
```

```python
def sliding_window_average(data, window_size):
    averages = []
    n = len(data)
    
    # Iterate over the data points
    for i in range(n):
        # Define the window boundaries
        first_window_item = max(0, i - window_size + 1)
        last_window_item = i
        window = data[first_window_item:last_window_item + 1]
        window_avg = sum(window) / len(window)
        averages.append(window_avg)  # Store the average
    return averages
```

```python
n=200_000
window_size=200
data=np.random.uniform(0, 1., size=(n,))
print(data)
```

```python
%%time
res=sliding_window_average(data, window_size)
print(res[0])
```
