# Brute Force Password O(2^N)

```python
from itertools import product, permutations # product is O(2^n) permutation is O(n!)

# Example usage of combinations from itertools
elements = ["0", "1"]

k = 4 # Generate all combinations of length k
```

```python
%%time

iter_res = product(elements, repeat=k)


# Output the combinations
count=0
print("Combinations:")
for c in iter_res: # time O(2**N), data O(1)
    count+=1
print("# of combinations : ",count)
```

# Generate all potential sequence

```python
sequence = range(11)  # The sequence of objects
```

```python
%%time

iter_res = permutations(sequence)

#list_res=list(iter_res) # O(n!) memory O(n!) time

# Output the combinations
count=0
print("Combinations:")
for c in iter_res: # O(1) O(n!) time
    count+=1
print("# of combinations : ", count)
```
