 # Merge sort

```python
# Merge two sorted subarrays into one sorted array
def merge(left, right):
    result = []
    i = j = 0
    
    # Compare elements from left and right subarrays and merge them in sorted order
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    # Append any remaining elements from the left or right subarray
    result.extend(left[i:])
    result.extend(right[j:])
    
    return result

# Recursive Merge Sort function
def merge_sort(arr):
    # Base case: an array with 1 or 0 elements is already sorted
    if len(arr) <= 1:
        return arr
    
    # Split the array into two halves
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    # Merge the two sorted halves
    return merge(left, right)
```

```python
import numpy as np
n=200_000
arr = np.random.randint(0, 100, (n,)) # data O(n)
print(arr)
```

```python
%%time

sorted_arr = merge_sort(arr)  # processing O(n*log(n))
print("Sorted array:", sorted_arr)
```
