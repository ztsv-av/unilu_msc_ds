# Matrix addition

```python
%%time

n=2000
matrixA = np.random.randint(0, 100, size=(n, n)) # O(n²)
matrixB = np.random.randint(0, 100, size=(n, n)) # O(n²)

# O(n²)

print("matrix A")
print(matrixA)

print("matrix B")
print(matrixB)
```

```python
%%time

result = np.zeros((n,n)) # initialize result matrix with zeros

# Add corresponding elements of matrix1 and matrix2
for i in range(n):
    for j in range(n):
        result[i][j] = matrixA[i][j] + matrixB[i][j]
print(result)
```

# Nearest Neighbors

```python
import numpy as np

def nearest_neighbors(data):
    for i in range(len(data)):
        nearest_dist=float('inf')
        nearest_neigh=None
        for j in range(len(data)):
            if i!=j :
                cur_dist=np.sqrt(np.sum((data[i] - data[j]) ** 2))
                if nearest_dist>cur_dist:
                    nearest_dist=cur_dist
                    nearest_neigh=data[j]
        print(f"The nearest neighbor of {data[i]} is {nearest_neigh} dist: {nearest_dist}")

        
```

```python
n=1000
data = np.random.randint(0, 100, size=(n, 2))
print(data)
```

```python
%%time


nearest_neighbors(data)

# Question: What happens to big O if we don't have 2 coordinates but `n` coordinates ? (high-dimension space)
```
