# Binary search

```python
def binary_search(arr, target):
    lower_bound = 0
    upper_bound = len(arr) - 1
    while lower_bound <= upper_bound:
        mid = lower_bound + (upper_bound - lower_bound) // 2  # Compute the middle among potential value

        # Check and update the possibility range
        if arr[mid] == target:
            return True
        elif arr[mid] > target:
            upper_bound = mid - 1
        else:
            lower_bound = mid + 1
    return False


sorted_array = [1, 3, 5, 7] # <- assume the array is always sorted
target_value = 1
result = binary_search(sorted_array, target_value) # O(log(n))


print(f"Element found ? {result}")

# Is it considered as fast ?
# What happens if `sorted_array` is very large (e.g. 10**10) ?
```

# Guess secret number O(log(n))

```python
def guess_secret_number(secret_number, lower_bound, upper_bound):
    # Initialize boundaries
    num_moves = 0
    continu = True

    while continu:
        mid = lower_bound + (upper_bound - lower_bound) // 2  # Calculate mid based on inverted bounds
        num_moves += 1
        print("AI is trying the number:", mid)

        # Check and update boundaries
        if secret_number == mid:
            print("Secret number found after", num_moves, "moves")
            continu = False  # exit
        elif secret_number < mid:
            upper_bound = mid  # Update upper_bound if the guess is too high
            print("Human says 'It is less!'")
        else:
            lower_bound = mid  # Update lower_bound if the guess is too low
            print("Human says 'It is more!'")

# Example usage
guess_secret_number(200005554466556556550, 0, 10**100)


# Notice 1: How to measure the computer power ? Memory usage ?
# Notice 2: What is the maximum number of moves ?
# Notice 3: Replace 100 with 10**100, is it slow ? Why ?
# Notice 4: It is necessary to use HPC for accelerating the computation ?
```
