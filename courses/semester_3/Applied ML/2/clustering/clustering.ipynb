# Clustering

Importing base libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## *Stars*

Loading a dataset, looking what's inside, and finding **3** clusters.

```python
stars = pd.read_csv("stars.csv")
print(stars.head())
```

```python
plt.scatter(stars.X, stars.Y)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, n_init="auto")
kmeans.fit(stars)
labels = kmeans.labels_

for i in np.unique(labels):
    cluster = stars[labels == i]
    plt.scatter(cluster.X, cluster.Y, label=f"cluster {i}")

plt.legend()
plt.plot()
```

## *Belts*

Load the file `belts.csv`, explore it, and cluster it!

Be careful of the choosen algorithm ;)

```python
# load the dataset
# print what's inside
belts = pd.read_csv("belts.csv")
belts
```

```python
# visualize it
plt.scatter(belts["X"], belts["Y"])
```

```python
# cluster it
from sklearn.cluster import DBSCAN

clustering_dbscan = DBSCAN(eps=0.2, min_samples=5).fit(belts)
labels_dbscan = clustering_dbscan.labels_

for i in np.unique(labels_dbscan):
    cluster = belts[labels_dbscan == i]
    plt.scatter(cluster.X, cluster.Y, label=f"cluster {i}")

plt.legend()
plt.plot()
```
